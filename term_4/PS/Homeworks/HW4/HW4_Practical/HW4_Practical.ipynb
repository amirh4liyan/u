# ****Question 2: Gaussian Blur (35 Points)****

In this problem, we will explore the concept of **Gaussian Blur**, a popular technique in image processing used to reduce noise and smooth out an image. Gaussian Blur relies on the **2D Gaussian (Multivariate Normal) distribution** and is implemented using **convolution**.

---

## **What is Gaussian Blur?**

Gaussian Blur works by smoothing an image through convolution with a **Gaussian kernel**. This kernel is derived from the 2D Gaussian distribution, which is defined as:

$$
f(x, y) = \frac{1}{2\pi \sigma^2} \exp\left(-\frac{x^2 + y^2}{2\sigma^2}\right)
$$

Where:
- $ \sigma $ is the standard deviation of the Gaussian distribution, which controls how "spread out" the blur is.
- $ x, y $ are the coordinates in the kernel.

The Gaussian kernel is a matrix of values computed from this function. When convolved with an image, it replaces each pixel with a weighted average of its neighboring pixels, where the weights are determined by the Gaussian kernel. This results in a smooth, blurred version of the original image.

---

## **Convolution in Gaussian Blur**

Convolution is the mathematical operation used to apply the Gaussian kernel to an image. For a 2D image \( I \) and a kernel \( K \), the convolution is defined as:

$$
(I * K)(x, y) = \sum_{i} \sum_{j} I(x-i, y-j) \cdot K(i, j)
$$

Where:
- $ I(x-i, y-j) $ is the pixel value at location $ (x-i, y-j) $ in the image.
- $ K(i, j) $ is the value at location $ (i, j) $ in the kernel.

In Gaussian Blur, the kernel $ K $ is the Gaussian kernel, and we perform this convolution on the image. Note that for images with 3 channels (RGB), we perform the convolution on each channel separately using the same Gaussian kernel.
